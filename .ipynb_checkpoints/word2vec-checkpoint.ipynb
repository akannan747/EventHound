```python
# imports needed and logging
import gensim, pandas as pd, re

event_data = pd.read_csv('event_data4.csv')
descriptions = []

for description in event_data[' description']:
    description = re.sub("[^a-zA-Z]", " ", description)
    description = description.lower().split()
    descriptions += [description]
```

```python
model = gensim.models.Word2Vec(descriptions, size=100, window=10, min_count=2, workers=10)  
model.train(descriptions, total_examples=len(descriptions), epochs=10)
```

```python
model.wv.most_similar(positive=['drink'])
```

```python
best_score = 0
best_description = None
query = 'party'
for description in descriptions:
    description_total = 0
    relevant_words = 0
    for word in description:
        try:
            word_score = model.wv.similarity(w1=query, w2=word)
        except:
            word_score = 0
        if word_score > .2:
            relevant_words += 1
        description_total += word_score
    if len(description) > 0:
        description_score = description_total*relevant_words/len(description)
    else:
        description_score = -1
    print(description_score)
    if description_score > best_score:
        best_score = description_score
        best_description = description
print(best_description, best_score)
```
